# DATA VISUALIZATION WITH PYTHON(Matplotlib/Seaborn)

```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
#import bikeshare dataset
df= pd.read_csv('bikeshare_data.csv')
df.head()
```

## Exploratory data analysis vs Explanatory data analysis

Exploratory analysis is carried out to discover insights. They are for your own consumption and do not necessarily have to be perfect. They help provide answers to analytical questions asked.


Explanatory analysis is carried out for the consumption of others. A story needs to be conveyed. It has to be easy to understand, insightful, and visually appealing.

**Matplotlib:** a versatile library for visualizations, but it can take some coding effort to put together common visualizations.
**Seaborn:** built on top of matplotlib, adds a number of functions to make common statistical visualizations easier to generate.

**Bad visuals**

-inaccurate results

### Data Types

For best output,it is important to know your data types as that determines the type of chart plotted.


Qualitative/categorical data

Nominal data: Data labels without any order

Ordinal data: data groups an order/ranking 


Quantitative or numeric types

Discrete quantitative data types usually have discrete values e.g whole numbers.

Continuous quantitative data types can be seen as further broken down discrete values e.g decimals.

**Matplotlib:** plt.plot(x,y), .scatter(x,y), .bar(x,height)

**Seaborn:** There are 2 main methods
1. sb.scatterplot(),sb.countplot, sb.lineplot(), sb.histplot() etc.


2. sb.relplot(data='', x='', y='',kind=''), sb.displot(), sb.catplot() etc.

No2 are figure level interfaces and make switching between different plot types possible by just altering the 'kind' parameter in most cases

```python
# color scheme for seaborn charts going forward

sb.set_style('darkgrid') # set grid color
sb.set_palette('Set2') # set color palette
```

### Univariate

### What is the ratio of subscribers to non subscribers ?

```python
# plotting a count chart
sb.countplot(data=df, x="User Type")

plt.title('Count of User Type')
plt.xlabel('User Type')
plt.ylabel('Count')

# removing the right and top borders of the axes
sb.despine()

plt.show()
```

### What day of the week had the highest bike rides?

```python
# plotting a count chart
sb.countplot(data=df, y="Weekday")

plt.title('Count per weekday')
plt.ylabel('Weekday')
plt.xlabel('Count')

# removing the right and top borders of the axes
sb.despine()

plt.show()
```

### What percentage of users were male and female?

```python
#plotting using plt.pie matplotlib

sex_val = df['Gender'].value_counts()

plt.figure(figsize=[10,8])

plt.pie(sex_val, startangle = 90, labels=sex_val.index, autopct='%.2f', counterclock = False, textprops={'fontsize': 20});

#  autopct='%.2f' - places the values in the sections
```

81% of users were male while about 19% of users were female

### What age distribution can be observed?

```python
#histogram with matplotlib
plt.hist(df.Age, bins=12);

plt.title('Age Distribution')
plt.ylabel('Count')
plt.xlabel('Age');
```

```python
#histogram with seaborn
sb.histplot(data=df, x="Age", bins=10)


plt.title('Age Distribution')
plt.ylabel('Count')
plt.xlabel('Age');
```

### Bivariate

```python
sb.boxplot(data=df, x='Gender', y='Age')
plt.xticks(rotation = 45);
```

```python
# In the sb.regplot() function below, the `truncate` argument accepts a boolean.
# If truncate=True, the regression line is bounded by the data limits. 

plt.figure(figsize=[8,6])
sb.regplot(data =df, x='Trip Duration', y='Age', truncate=True);
```

```python
plt.figure(figsize=[8,6])
sb.countplot(data=df, x='User Type', hue='Gender')
```

### Multivariate

```python
sb.boxplot(data=df, x='Weekday', y='Age', hue='Gender')
plt.xticks(rotation = 45);
```
